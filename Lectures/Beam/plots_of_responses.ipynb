```python
import numpy as np
import csv
import matplotlib.pyplot as plt
```

```python
def csv2list(input,col_list):
    time = []
    setpoint= []
    response = []
    voltage = []
    with open(input,'r') as csvfile:
        next(csvfile, None)
        plots = csv.reader(csvfile, delimiter=',')
        for row in plots:
            time.append(float(row[col_list[0]]))
            setpoint.append(float(row[col_list[1]]))
            response.append(float(row[col_list[2]]))
            voltage.append(float(row[col_list[3]]))
    return time,setpoint,response,voltage
```

```python
files = ["30to45_cleaned.csv","45to60_cleaned.csv","60to30_cleaned.csv"]
```

```python
for file in files:
    h_t,h_sp,h_r,h_v = csv2list(file,range(0,4))
    j_t,j_sp,j_r,j_v = csv2list(file,range(4,8))
    d_t,d_sp,d_r,d_v = csv2list(file,range(8,12))

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16,6))
    ax1.plot(h_t,h_sp, label='Setpoint', color="black", linestyle="dashed") 
    ax1.plot(h_t,h_r, label='Harry', color="orange")
    ax1.plot(j_t,j_r, label='Jack', color="green")
    ax1.plot(d_t,d_r, label='Daniel', color="blue")
    ax1.set_xlabel("Time [s]")
    ax1.set_ylabel(u"Position []")
    ax1.set_title("Position Control with kp=3.4 ki=0.2 kd=0.65")
    ax1.legend(loc=1)

    ax2.plot(h_t,h_v, label='Harry', color="orange")
    ax2.plot(j_t,j_v, label='Jack', color="green")
    ax2.plot(d_t,d_v, label='Daniel', color="blue")
    ax2.set_xlabel("Time [s]")
    ax2.set_ylabel("Voltage [V]")
    ax2.set_title("Voltage vs Time")
    ax2.legend(loc=1)
    
    plt.savefig(file[:-4]+".png")


```

```python
sine_freq = ["0_015_Hz.csv", "0_1_Hz.csv", "1_Hz.csv"]
freq = ["0.015", "0.1", "1"]
```

```python
i = 0
for file in sine_freq:
    h_t,h_sp,h_r,h_v = csv2list(file,range(0,4))

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16,6))
    ax1.plot(h_t,h_sp, label='Setpoint', color="black", linestyle="dashed") 
    ax1.plot(h_t,h_r, label='Response', color="blue")
    ax1.set_xlabel("Time [s]")
    ax1.set_ylabel(u"Position []")
    ax1.set_title("Position Control with A = 20, f ="+ freq[i] + "Hz Offset = 45")
    ax1.legend(loc=1)

    ax2.plot(h_t,h_v, color="blue")
    ax2.set_xlabel("Time [s]")
    ax2.set_ylabel("Voltage [V]")
    ax2.set_title("Voltage vs Time")
    ax2.legend(loc=1)
    
    i = i+1
    plt.savefig(file[:-4]+".png")


```
